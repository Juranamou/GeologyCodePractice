```python
import numpy as np
import matplotlib.pyplot as plt

#X values 0 to +1, 100 of them
x = np.linspace(0, 1.0, 100)
#Y values 0 to +1, 100 of them
y = np.linspace(0, 1.0, 100)

#Defines the cordinate space
X, Y = np.meshgrid(x,y)

#Look does this make sense? I am printing the maximum value of meshgrid
print(X[1,99])

#Look its making 100 x values for all 100 y values.. Cool!
print(np.shape(X))

# radius of the circle
r = 0.1

# theta goes from 0 to 2pi for a 100 values
theta = np.linspace(0, 2*np.pi, 100)

# make x and y
x = r*np.cos(theta)
y = r*np.sin(theta)

# create the figure
fig, ax = plt.subplots(1)
ax.plot(x,y)
plt.show()

```
