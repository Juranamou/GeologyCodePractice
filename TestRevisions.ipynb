## Linear Algebra

### Exercise 11 (8 points)
Using the linear algebra module linalg previously imported, calculate the **norm** and the **determinant** of random matrixes of increasing size, from $1x1$ up to $100x100$ and plot the result in function of the matrix size. It is clearer if you plot np.log10() of the determinant vs the linear plot of the norm. The determiant can be negative, so you can plot the log of the absolute value.

```python
# here write your solution

OneByOne = np.random.rand(1,1)
TenByTen = np.random.rand(10,10)

NormOne=linalg.norm(OneByOne)
NormTen=linalg.norm(TenByTen)
Normals=np.array([NormOne,NormTen])

DetOne=linalg.det(OneByOne)
DetOne=np.abs(DetOne)
DetTen=linalg.det(TenByTen)
DetTen=np.abs(DetTen)

LogOne=np.log10(DetOne)
LogTen=np.log10(DetTen)
Logs=np.array([LogOne,LogTen])

plt.scatter(Normals,Logs)
```
